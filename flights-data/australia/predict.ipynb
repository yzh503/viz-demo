```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('intl-traffic.csv', parse_dates=['date'])
# sort
data.sort_values(by=['date'], inplace=True)
data.fillna(method='ffill', inplace=True) # forward fill
data.fillna(method='bfill', inplace=True)
```

```python
predictions = []
error_list = []

for city in data['name'].unique():
    city_data = data[data['name'] == city].sort_values(by='date').set_index('date')
    
    # SARIMA model
    model = SARIMAX(city_data['value'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
    model_fit = model.fit(disp=False)
    
    # Prediction until 2024-12-01
    num_steps = (pd.to_datetime('2024-12-01') - city_data.index[-1]).days // 30
    forecast = model_fit.get_forecast(steps=num_steps)
    predicted_index = [city_data.index[-1] + pd.DateOffset(months=i+1) for i in range(num_steps)]
    
    # Plot
    plt.figure(figsize=(10, 2))
    city_data['value'].plot(label='Actual', legend=True)
    plt.plot(predicted_index, forecast.predicted_mean, label='Forecast', color='red')
    plt.title(city)
    plt.legend()
    plt.show()
    
    # Append to predictions list
    for date, pred in zip(predicted_index, forecast.predicted_mean):
        predictions.append((city, date, pred))
    
    # Error measurement
    y_true = city_data['value'].values
    y_pred = model_fit.fittedvalues.values
    error = mean_squared_error(y_true, y_pred, squared=False)
    error_list.append((city, error))

    plt.show()

# Convert predictions to DataFrame
predictions_df = pd.DataFrame(predictions, columns=['name', 'date', 'value'])

# Append predictions to original data
full_data = pd.concat([data, predictions_df], axis=0).reset_index(drop=True)
```

```python

full_data['value'] = full_data['value'].astype(int)
full_data = full_data[full_data['date'] <= pd.to_datetime('2024-12-01')]
full_data.to_csv('intl-traffic-predictions.csv', index=False)
```
